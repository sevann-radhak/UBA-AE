```python
import numpy as np
import matplotlib.pyplot as plt
from pyswarm import pso

def objective_function(x):
    return -np.exp(-0.1 * (x[0]**2 + x[1]**2)) * np.cos(x[0]) * np.sin(x[1])

lb = [-50, -50]
ub = [50, 50]

xopt, fopt = pso(objective_function, lb, ub)

max_value = -fopt

x = np.linspace(-50, 50, 400)
y = np.linspace(-50, 50, 400)
X, Y = np.meshgrid(x, y)
Z = np.exp(-0.1 * (X**2 + Y**2)) * np.cos(X) * np.sin(Y)

plt.figure(figsize=(10, 8))
plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar(label='z(x, y)')
plt.scatter(xopt[0], xopt[1], color='green', label='Optimal Point', s=100)
plt.title('Objective Function z(x, y)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()

print("Optimal Solution (x, y):", xopt)
print("Optimal Value (z):", max_value)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from pyswarm import pso

gbest_values = []

def objective_function_with_gbest(x):
    value = objective_function(x)
    gbest_values.append(-value)  
    return value

xopt, fopt = pso(objective_function_with_gbest, lb, ub, swarmsize=100, maxiter=100)

plt.figure(figsize=(10, 6))
plt.plot(gbest_values, label='gbest over iterations')
plt.title('Global Best Value Over Iterations')
plt.xlabel('Iterations')
plt.ylabel('gbest Value')
plt.legend()
plt.grid()
plt.show()

max_value = -fopt
print("Optimal Solution (x, y):", xopt)
print("Optimal Value (z):", max_value)
```
